```python
import pandas as pd

# Ler o arquivo de planilha do Excel
df = pd.read_excel('PINH04900_2012_2017.xlsx')
df2 = pd.read_excel('PINH04900_2017_2022.xlsx')

# Converter a coluna "Data Coleta" para o formato de data
df['Data Coleta'] = pd.to_datetime(df['Data Coleta'], format='%d/%m/%Y')
df2['Data Coleta'] = pd.to_datetime(df2['Data Coleta'], format='%d/%m/%Y')
df2['Graficos Digitalizados'] = pd.to_datetime(df2['Data Coleta'], format='%d/%m/%Y')

# Filtrar as linhas para os anos de 2012 e 2013
df_f0 = df[(df['Data Coleta'].dt.year == 2012)]
df_f1 = df[(df['Data Coleta'].dt.year == 2013) & (df['Data Coleta'].dt.month == 5)]
df_f2 = df[(df['Data Coleta'].dt.year == 2014) & (df['Data Coleta'].dt.month == 5)]
df_f3 = df[(df['Data Coleta'].dt.year == 2015) & (df['Data Coleta'].dt.month == 5)]
df_f4 = df[(df['Data Coleta'].dt.year == 2016) & (df['Data Coleta'].dt.month == 5)]
df_f5 = df[(df['Data Coleta'].dt.year == 2017) & (df['Data Coleta'].dt.month == 5)]
df_f51 = df2[(df2['Data Coleta'].dt.year == 2017) & (df2['Data Coleta'].dt.month == 5)]
df_f6 = df2[(df2['Data Coleta'].dt.year == 2018) & (df2['Data Coleta'].dt.month == 5)]
df_f7 = df2[(df2['Data Coleta'].dt.year == 2019) & (df2['Data Coleta'].dt.month == 5)]
df_f8 = df2[(df2['Data Coleta'].dt.year == 2020) & (df2['Data Coleta'].dt.month == 5)]
df_f9 = df2[(df2['Data Coleta'].dt.year == 2021) & (df2['Data Coleta'].dt.month == 5)]
df_f10 = df2[(df2['Data Coleta'].dt.year == 2022) & (df2['Data Coleta'].dt.month == 5)]

parametros_desejados = [
    'Turbidez',
    'Nitrogênio Total',
    'Nitrogênio-Nitrito',
    'Nitrogênio-Nitrato',
    'Nitrogênio Amoniacal',
    'Nitrogênio Kjeldahl',
    'Oxigênio Dissolvido',
    'DBO (5, 20)',
    'Sólido Total',
    'Fósforo Total',
    'pH',
    'Temperatura da Água',
    'Temperatura do Ar',
    'Escherichia coli**'
]

df_f0 = df_f0[df_f0['Parametro'].isin(parametros_desejados)]
df_f1 = df_f1[df_f1['Parametro'].isin(parametros_desejados)]
df_f2 = df_f2[df_f2['Parametro'].isin(parametros_desejados)]
df_f3 = df_f3[df_f3['Parametro'].isin(parametros_desejados)]
df_f4 = df_f4[df_f4['Parametro'].isin(parametros_desejados)]
df_f5 = df_f5[df_f5['Parametro'].isin(parametros_desejados)]
df_f51 = df_f51[df_f51['Parametro'].isin(parametros_desejados)]
df_f6 = df_f6[df_f6['Parametro'].isin(parametros_desejados)]
df_f7 = df_f7[df_f7['Parametro'].isin(parametros_desejados)]
df_f8 = df_f8[df_f8['Parametro'].isin(parametros_desejados)]
df_f9 = df_f9[df_f9['Parametro'].isin(parametros_desejados)]
df_f10 = df_f10[df_f10['Parametro'].isin(parametros_desejados)]

dfs = {'2012' : df_f0, 
       '2013' : df_f1, 
       '2014' : df_f2,
       '2015' : df_f3, 
       '2016' : df_f4,
       '2017' : df_f5,
       '2018' : df_f6,
       '2019' : df_f7,
       '2020' : df_f8,
       '2021' : df_f9,
       '2022' : df_f10
}

# for ano, df in dfs.items():
#     dfs[ano].to_excel('arquivo_' + ano + '-NOV.xlsx', index=False)
```

```python
for ano, df in dfs.items():
    df['Valor'] = df['Valor'].astype(str)
    df['Valor'] = df['Valor'].str.replace(',', '.')
    df['Valor'] = df['Valor'].astype(float)

    media_turbidez = df[df['Parametro'] == 'Turbidez']['Valor'].mean()
    
    n1 = df[df['Parametro'] == 'Nitrogênio-Nitrito']['Valor'].mean()
    n2 = df[df['Parametro'] == 'Nitrogênio-Nitrato']['Valor'].mean()
    n4 = df[df['Parametro'] == 'Nitrogênio Kjeldahl']['Valor'].mean()
    nitrogenio_total_soma = (n1 + n2 + n4)/3
    nitrogenio_total = df[df['Parametro'] == 'Nitrogênio Total']['Valor'].mean()

    o2_dissolvido = df[df['Parametro'] == 'Oxigênio Dissolvido']['Valor'].mean()
    
    dbo = df[df['Parametro'] == 'DBO (5, 20)']['Valor'].mean()
    
    solido_total = df[df['Parametro'] == 'Sólido Total']['Valor'].mean()
    
    fosforo_total = df[df['Parametro'] == 'Fósforo Total']['Valor'].mean()
    
    ph = df[df['Parametro'] == 'pH']['Valor'].mean()
    
    t_agua = df[df['Parametro'] == 'Temperatura da Água']['Valor'].mean()
    
    t_ar = df[df['Parametro'] == 'Temperatura do Ar']['Valor'].mean()
    
    coliformes = df[df['Parametro'] == 'Escherichia coli**']['Valor'].mean()
    
    print("ANO: ", ano)
    print("Fosforo-Total: ", round(fosforo_total, 3))
    print("pH: ", round(ph, 3))
    print("DBO: ", round(dbo, 3))
    print("Nitrogênio Total: ", round(nitrogenio_total, 3))
    print("NTOTALSOMA: ", round(nitrogenio_total_soma, 3))
    print("Colifomres: ", round(coliformes, 3))
    #print("T-agua: ", t_agua)
    #print("T-ar: ", t_ar)
    print("Temperatura", round(t_agua - t_ar, 3))
    print("Turbidez: ", round(media_turbidez, 3))
    print("Solido-Total: ", round(solido_total, 3))
    print("Oxigênio Dissolvido: ", round(o2_dissolvido, 3), "\n")
```
